## Import Dependencies

```python
#Import Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

## Data Acquisition

```python
# Load the Iris dataset into DataFrame
df = sns.load_dataset('iris')

# Inspect first rows of the DataFrame
print(df.head())
```

The first rows of the Iris dataset data are displayed above.

## Initial Data Inspection

```python
print(df.head(10))
print(df.shape)
print(df.dtypes)
```

The first 10 rows, rows and columns, and data types in the Iris Dataset are displayed above.

## Initial Descriptive Statistics

```python
print(df.describe())
```

Summary statistics for each column of the Iris Dataset are shown above.

## Creating a Histogram to display variables in the Iris Dataset

```python
# Inspect histogram by numerical column
df['sepal_length'].hist()

# Inspect histograms for all numerical columns
df.hist()

# Show all plots
plt.show()
```

**Conclusion**: Using a histogram to display sepal length, sepal width, petal length, and petal width I observed that each measurement reveals different patterns and distributions that will help distinguish between the species in the Iris dataset. 

## Displaying the counts of each species with plots

```python
# Inspect value counts by categorical column
df['species'].value_counts()

# Inspect value counts for all categorical columns
for col in df.select_dtypes(include=['object', 'category']).columns:
    # Display count plot
    sns.countplot(x=col, data=df)
    plt.title(f'Distribution of {col}')
    plt.show()

# Show all plots
plt.show()
```

**setosa count**: 50

**versicolor count**: 50

**virginica count**: 50


**Conclusion**: Each species of flower in the Iris dataset has an equal representation of 50.

## Performing Transformation on the columns in the Iris Dataset

```python
# Renaming a column
df.rename(columns={'sepal_length': 'Sepal Length', 'sepal_width': 'Sepal Width'}, inplace=True)

# Adding a new column
df['Sepal Dimensions'] = df['Sepal Length'] * df['Sepal Width']

# Print the head and new column
print(df.head())
```

## Creating a Pairplot to identify trends from the Iris Dataset Species

```python
# Pairplot
sns.pairplot(df, hue='species')
plt.show()
```

**Conclusion**: The pairplot above shows scatter plots for variables: Sepal Length, Sepal Width, Petal Length, Petal Width, Sepal Area and Sepal Dimensions. The species of flowers are highlighted in different colors, which are used to distinguish between the variables and identify trends and differences.

My main takeaway from the pairplot was how different the Setosa variables were compared to the other two species. Also, there wasn't much difference between the petal length and petal width. When I looked at the two charts they were hard to differentiate indicating strong correlation between the petal length and petal width.

## Creating a Box Plot to visualize the petal width distribution by species

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x='species', y='petal_width', data=df)
plt.title('Petal width Distribution by Species')
plt.show()
```

**Conclusion**: It's very easy to differentiate between the petal lengths of the three species by looking at the box plot. Details of each species findings below.

**Setosa**: The shortest petal width out of the three species.

**Versicolor**: The average or middle petal width between the three species.

**Virginica**: The largest petal width of the three species.

## Creating a Box Plot to visualize the petal length distribution by species

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x='species', y='petal_length', data=df)
plt.title('Petal length Distribution by Species')
plt.show()
```

**Conclusion**: Same as the first box plot it is easy to distinguish between the species petal length using the box plot.

**Setosa**: The shortest petal lengths out of the three species by a wide margin. 

**Versicolor**: The middle petal length of the species, but closer to virginica's length than setosa's.

**Virginica**: The largest petal lengths of the three species.
